#GANについて

##GAN総論

```python
display_png(Image('GANs.png'))
```



**GAN (Generative Adversarial Network)**では2種類のネットワークを用意する.  
1. Generator : 画像を生成するネットワーク  
2. Discriminator : Gから生成された画像が偽画像か/訓練データ画像かを分類するネットワーク  



GはDを騙そうと訓練データに近い画像を生成するよう学習する.  
DはGに騙されないよう学習する.  
このように「DとGをバトらせて,　Gの生成する画像を訓練データのような現実の画像に近づける」ことがGANの目的.

##GANの理論

$p_z$ : 乱数zの分布, $p_g$ : Gの生成するデータの分布, $p_r$ : 現実のデータの分布  
$G(x)$ : Generatorが生成するデータ  
$D(x)$ : Discriminatorがxを訓練データであると識別する確率

Dは, 訓練データを訓練データとして正しく識別させたい. つまり次式を最大化したい. $$ \mathbb{E} _ {x \sim p _ r(x)} \left[ \log D(x) \right] $$  一方, DにはGが生成したデータ G(z)は正しく偽物と識別して欲しい. よって次式を最大化したい. $$ \mathbb{E} _ {z \sim p _ z(z)} \left[ \log \left( 1 - D(G(z) \right) \right] $$


Gは, 生成したデータをDに本物だと判断させたい. つまり, 次式を最小化したい. $$ \mathbb{E} _ {z \sim p _ z(z)} \left[ \log \left( 1 - D(G(z) \right) \right] $$

以上をまとめると, GANのmin-max lossは, 
$$ \min_G \max_D L(D,G) = \mathbb{E} _ {x \sim p _ r(x)} \left[ \log D(x) \right] + \mathbb{E} _ {z \sim p _ z(z)} \left[ \log \left( 1 - D(G(z) \right) \right] \\ = \mathbb{E} _ {x \sim p _ r(x)} \left[ \log D(x) \right] + \mathbb{E} _ {x \sim p _ g(x)} \left[ \log \left( 1 - D(x) \right) \right] $$

・Discriminatorの最適解をまず求める. $L(G,D)$の期待値の部分は次のように書ける.

$$ L(G,D) = \int \left( p _ {r} (x) \log(D(x)) + p _ {g}(x) \log(1 - D(x)) \right) dx $$

$L(G,D)$を最大化するような$D(x)$が求めるDiscriminatorの最適解である. 

$$ \hat{x}=D(x), A = p _ r(x), B = p _ {g}(x) $$とおくと, 


$$ f(\hat{x}) = A\log \hat{x} + B \log (1- \hat{x}) $$ $x̂ $について微分すれば

$$ \frac{d f(\hat{x})}{d\hat{x}} = \frac{A-(A+B)\hat{x}} {\hat{x} (1- \hat{x})} $$ これを0とおくと, 最適な$D(x)$は

$$ D^{\ast}(x) = \frac{A}{A+B} = \frac{ p _ r(x)} { p _ r(x)+p _ {g}(x)} $$

Generatorが最適に学習すると, $p_g$は$p_r$に近づいていく.  
$p_g=p_r$のとき $D∗(x)=1/2$であり, 完璧なGeneratorができると、Discriminatorは機能しなくなる（五分五分で推論してしまう）

・次に, Generatorの最適解を求める. 

\begin{align}
L(G,D^\ast)&=\int \left( p_r\log \frac{p_r}{p_g+p_r}+p_g \log \frac{p_g}{p_g+p_r} \right) dx \\
&= \int \left\{ p_r \log \left( \frac{1}{2}\frac{p_r}{\frac{p_g+p_r}{2}} \right) + p_g \log \left( \frac{1}{2}\frac{p_g}{\frac{p_g+p_r}{2}} \right) \right\} dx \\
&= \int p_r \log \frac{p_r}{\frac{p_g+p_r}{2}} dx + \int p_g \log \frac{p_g}{\frac{p_g+p_r}{2}}dx - \int (p_r+p_g)\log 2 dx \\
&= D_{KL}\left( p_r \left| \left| \frac{p_g+p_r}{2} \right. \right.  \right) + D_{KL}\left( p_g \left| \left| \frac{p_g+p_r}{2} \right. \right.  \right) -\log 4 \\
&= 2D_{JS}\left( p_r || p_g \right) -\log 4
\end{align}

である. $p_r=p_g$のとき, 損失関数$L(G,D)$は最小値 $-\log4$をとる.

 （**Jensen–Shannon Divergence**）

$$D _ {JS}(p||q) = \frac{1}{2} D _ {KL} \left( p||\frac{p+q}{2} \right) +  \frac{1}{2} D _ {KL} \left( q||\frac{p+q}{2} \right)$$


##GANの損失関数の実装について

実装においては, 上述のようなJensen-Shannon Divergenceの最適化を行うような手法はあまり取られず, 単純にクロスエントロピー(+シグモイド関数出力)で損失関数の計算を行うようなものが多い.



・Discriminatorの損失関数について  


入力データを$x$, Dの出力を$y=D(x)$, ラベルを$l(l=\{0,1\})$とすると, ミニバッチのデータM個分の判別の同時確率は
$$\prod_{i=1}^M y_i^{l_i}(1-y_i)^{1-l_i}$$
であり, これを最大化するようDを学習させたい. 対数をとり, マイナスをかけると, Discriminatorの損失関数は$$-\sum_{i=1}^M[l_i\log y_i +(1-l_i)\log(1-y_i)]$$


・Generatorの損失関数について

Generatorは, Dを騙したいので, Dの損失関数
$$-\sum_{i=1}^M[l_i\log y_i +(1-l_i)\log(1-y_i)]$$
を最大化すればよい(↔︎Dは損失関数を最小化しようとしている). よって, Gの損失関数は$$\sum_{i=1}^M[l_i\log y_i +(1-l_i)\log(1-y_i)]$$

Gの生成した偽画像では$l_i=0$であり, 入力$z_i$と出力$y=D(x)$で式を書き直すと, $$ \sum_{i=1}^M \log (1-D(G(z_i)))$$

これがGの損失関数である.

しかし, この式だとGの学習が進みにくいことが知られている. 初期のGが生成する画像は教師データとの違いが大きすぎて, 学習の進んでいないDでも正しく判別できてしまう. 正しく判別されるとき$D(G(z_i))=0$だから, Gの損失関数で $\log (1-D(G(z_i)))=0$ となって学習が進まない.

そこで, Gの損失関数としてかわりに次式を用いる.
$$-\sum_{i=1}^M \log D(G(z_i))$$
この式だと, GがDを上手く騙した時($D(G(z_i))=1$)は損失関数は０となる. また, GがDを騙せなかったとき, $\log $の中身は0~1の値をとり, 損失関数は正の大きな値をとることになる.

[要するに]  
前者の損失関数ではDに見抜かれても損失は０で学習が進まない  
後者の損失関数ではDに見抜かれると損失がやばい↑↑



*これらの損失関数は, torch.nn.BCEWithLogitsLoss()で書ける (実装は学習の実装の項で) 

https://pytorch.org/docs/master/generated/torch.nn.BCEWithLogitsLoss.html

```python
display_png(Image('TransposedConvolution.png'))
```

DCGANでは転置畳み込みを使う. 

**転置畳み込み**(**Transposed Convolution**)とは：  
・カーネルを入力データの1セルごとに計算する  
・1セルごとにカーネルとの掛け算を計算して、最後に全ての結果を足す

転置畳み込みを繰り返して特徴量マップの要素数を増やすことにより, 入力乱数から求めるサイズの画像が得られる.

PyTorchでは torch.nn.ConvTranspose2d 
https://pytorch.org/docs/master/generated/torch.nn.ConvTranspose2d.html#torch.nn.ConvTranspose2d

*普通の畳み込みでは、「入力データの複数のセル」とカーネルを対応させるが, 転置畳み込みでは, 「入力データの一つのセル」とカーネルを対応させる点が異なる.

##Generatorの実装

```python
class Generator(nn.Module):

    def __init__(self, z_dim=20, image_size = 64):
        super(Generator, self).__init__()

        self.layer1 = nn.Sequential(
            nn.ConvTranspose2d(z_dim, image_size*8, kernel_size=4, stride=1),
            nn.BatchNorm2d(image_size*8),
            nn.ReLU(inplace=True)
        )

        self.layer2 = nn.Sequential(
            nn.ConvTranspose2d(image_size*8, image_size*4, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(image_size*4),
            nn.ReLU(inplace=True)
        )

        self.layer3 = nn.Sequential(
            nn.ConvTranspose2d(image_size*4, image_size*2, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(image_size*2),
            nn.ReLU(inplace=True)
        )

        self.layer4 = nn.Sequential(
            nn.ConvTranspose2d(image_size*2, image_size, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(image_size),
            nn.ReLU(inplace=True)
        )

        self.last = nn.Sequential(
            nn.ConvTranspose2d(image_size, 1, kernel_size=4, stride=2, padding=1),
            nn.Tanh()
        )

    def forward(self,z):
        out = self.layer1(z)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = self.last(out)

        return out
```

```python
# 動作確認
import matplotlib.pyplot as plt
%matplotlib inline

G = Generator(z_dim=20, image_size=64)

# 入力する乱数
input_z = torch.randn(1, 20)
input_z = input_z.view(input_z.size(0), input_z.size(1), 1, 1) #Tensorを(1,20,1,1)に変形

fake_images = G(input_z)

img_transformed = fake_images[0][0].detach().numpy()
plt.imshow(img_transformed, 'gray')
plt.show()
```

#Discriminator

##Discriminatorのメカニズム

Dは（単なる）画像識別のネットワーク　 :　  
入力された画像を教師データと判断したら1, Gが生成した画像と判断したら0を出力する

・GANのDの活性化関数には**LeakyReLU**を用いる.  ( x<0 なら αx を出力するReLU)  
　D→Gへと誤差を逆伝播させて学習するが, DにReLUを用いてbackpropが0になってしまうとGの学習が上手く進まなくなってしまう.


##Discriminatorの実装

```python
class Discriminator(nn.Module):

    def __init__(self, z_dim=20, image_size=64):
        super(Discriminator, self).__init__()

        self.layer1 = nn.Sequential(
            nn.Conv2d(1, image_size, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.1, inplace=True) # GANのDにはLeakyReLUをつかう
        )

        self.layer2 = nn.Sequential(
            nn.Conv2d(image_size, image_size*2, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.1, inplace=True) # GANのDにはLeakyReLUをつかう
        )

        self.layer3 = nn.Sequential(
            nn.Conv2d(image_size*2, image_size*4, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.1, inplace=True) # GANのDにはLeakyReLUをつかう
        )

        self.layer4 = nn.Sequential(
            nn.Conv2d(image_size*4, image_size*8, kernel_size=4, stride=2, padding=1),
            nn.LeakyReLU(0.1, inplace=True) # GANのDにはLeakyReLUをつかう
        )

        self.last = nn.Conv2d(image_size*8, 1, kernel_size=4, stride=1)

    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = self.last(out)

        return out
```

```python
#動作確認
D = Discriminator(z_dim=20, image_size=64)

i#入力する乱数
input_z = torch.randn(1, 20)
input_z = input_z.view(input_z.size(0), input_z.size(1), 1, 1) #Tensorを(1,20,1,1)に変形

#Generatorで偽画像を生成
fake_images = G(input_z)

#Discriminatorに偽画像を入力
d_out = D(fake_images)

print(nn.Sigmoid()(d_out))
```

```python
# train, val の画像データとアノテーションデータへのファイルパスのリストを作成
def make_datapath_list():

    train_img_list = list()

    for img_idx in range(200):
        # 7の画像へのパスを格納
        img_path ='/content/data/img_78/img_7_' + str(img_idx)+ '.jpg'
        train_img_list.append(img_path)

        # 8の画像へのパスを格納
        img_path ='/content/data/img_78/img_8_' + str(img_idx)+ '.jpg'
        train_img_list.append(img_path)

    return train_img_list
```

```python
# 画像の前処理を行うクラス
class ImageTransform():

    def __init__(self, mean, std):
        self.data_transform = transforms.Compose([
                                                  transforms.ToTensor(),
                                                  transforms.Normalize(mean, std)
        ])

    def __call__(self, img):
        return self.data_transform(img)
```

```python
# 画像のDatasetクラス
class GAN_Img_Dataset(data.Dataset):

    def __init__(self, file_list, transform):
        self.file_list = file_list
        self.transform = transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self,index):
        img_path = self.file_list[index]
        img = Image.open(img_path) # HWC
        img_transformed = self.transform(img)

        return img_transformed # 画像→前処理→Tensorに変換→return
```

```python
# Datasetの作成
train_img_list=make_datapath_list()

mean = (0.5,)
std = (0.5,)
train_dataset = GAN_Img_Dataset(file_list=train_img_list, transform=ImageTransform(mean, std))


# DataLoaderの作成
batch_size = 64

train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# 動作確認
batch_iterator = iter(train_dataloader)
imges = next(batch_iterator)
print(imges.size()) # ([batch_size, channel, Height, Width])
```

#学習の実装

##ネットワークの初期化

Conv2d, ConvTranspose2dの重み　~ $N(0, 0.02)$  
BatchNorm2dの重み ~ $N(1, 0.02)$  
$bias = 0$

と初期化しておくと, DCGANでは上手くいくことが経験的に知られている.

```python
# ネットワークを初期化する関数
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

```python
# 初期化を行う
G.apply(weights_init)
D.apply(weights_init)
```

##学習の関数の実装

```python
def train_model(G, D, dataloader, num_epochs):

    # GPUが使えるか確認
    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    print('使用デバイス : ', device)

    # 最適化手法の設定
    g_lr, d_lr = 0.0001, 0.0004
    beta1, beta2 = 0.0, 0.9
    g_optimizer = torch.optim.Adam(G.parameters(), g_lr, [beta1, beta2])
    d_optimizer = torch.optim.Adam(D.parameters(), d_lr, [beta1, beta2])

    # 誤差関数の定義
    criterion = nn.BCEWithLogitsLoss(reduction='mean')

    # パラメータをハードコーディング
    z_dim = 20
    mini_batch_size = 64

    # ネットワークをGPUへ送る
    G.to(device)
    D.to(device)

    # ネットワークを訓練モードにする
    G.train()
    D.train()

    # これを書いておくと, ネットワークがある程度固定である時に計算が高速になる
    torch.backends.cudnn.benchmark = True

    # 画像の枚数
    num_train_imgs = len(dataloader.dataset)
    batch_size = dataloader.batch_size

    # イテレーションのカウンタをセット
    iteration = 1
    logs = []

    # epochのloop
    for epoch in range(num_epochs):

        t_epoch_start = time.time() #　開始時間の保存
        epoch_g_loss = 0.0
        epoch_d_loss = 0.0

        print('-----------')
        print('Epoch {}/{}'.format(epoch, num_epochs))
        print('-----------')
        print('(train)')

        # DataLoaderからミニバッチを取り出すループ
        for imges in dataloader:

            #---------------------------
            # Discriminatorの学習
            #---------------------------

            if imges.size()[0] ==1:
                continue # ミニバッチのサイズが1だとBatchNormでエラーになるので避けるようにする
            
            # GPUが使えたらGPUにデータを送る
            imges = imges.to(device)

            # 正解/偽ラベルを作成する
            mini_batch_size = imges.size()[0]
            label_real = torch.full((mini_batch_size,), 1).to(device) # Returns a tensor of size filled with fill_value
            label_fake = torch.full((mini_batch_size,), 0).to(device)

            # 真の画像を判定
            d_out_real = D(imges)

            # 偽の画像を生成して判定
            input_z = torch.randn(mini_batch_size, z_dim).to(device)
            input_z = input_z.view(input_z.size(0), input_z.size(1), 1, 1)
            fake_images = G(input_z)
            d_out_fake = D(fake_images)

            # 誤差の計算
            d_loss_real = criterion(d_out_real.view(-1), label_real) # criterion = nn.BCEWithLogitsLoss(reduction='mean')
            d_loss_fake = criterion(d_out_fake.view(-1), label_fake)
            d_loss = d_loss_real + d_loss_fake

            # Backprop
            g_optimizer.zero_grad()
            d_optimizer.zero_grad()
            
            d_loss.backward(retain_graph=True)
            d_optimizer.step()

            #-----------------------
            # Generatorの学習
            #-----------------------

            # 偽の画像を生成して判定
            input_z = torch.randn(mini_batch_size, z_dim).to(device)
            input_z = input_z.view(input_z.size(0), input_z.size(1), 1, 1)
            fake_imges = G(input_z)
            d_out_fake = D(fake_images)

            # 誤差の計算
            g_loss = criterion(d_out_fake.view(-1), label_real)

            # Backprop
            g_optimizer.zero_grad()
            d_optimizer.zero_grad()
            g_loss.backward(retain_graph=True)
            g_optimizer.step()

            #-----
            # 記録
            #-----

            epoch_d_loss += d_loss.item()
            epoch_g_loss += g_loss.item()
            iteration +=1

        # epochのphaseごとのlossとaccuracy
        t_epoch_finish = time.time()
        print('---------------')
        print('epoch{} || Epoch_D_Loss:{:.4f} || Epoch_G_LOss:{:.4f}'.format(
            epoch, epoch_d_loss/batch_size, epoch_g_loss/batch_size))
        print('timer: {:.4f} sec.'.format(t_epoch_finish - t_epoch_start))
        t_epoch_start = time.time()

    return G, D
```

```python
# 学習と検証の実行
num_epochs = 200
G_update, D_update = train_model(G, D, dataloader=train_dataloader, num_epochs=num_epochs)
```

epoch数を増やしすぎる(今回は1000)と, 7の画像しか生成しなくなる.

7の方が8より単純なために, Dを騙す画像を生成しやすく, Gが7ばかり生成してしまうようになる.

このように「教師データの一部しか生成しなくなる現象」を**モード崩壊**という.
